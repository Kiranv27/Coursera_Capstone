### import necessary libraries

```python
import pandas as pd
from bs4 import BeautifulSoup

inputhtml = "List of postal codes of Canada_ M - Wikipedia.html"
```

### read the input html: copy from data the 'List of postal codes of Canada_ M - Wikipedia' folder and html to notebook dir

```python
with open(inputhtml, encoding='utf8') as fp:
    soup = BeautifulSoup(fp, 'lxml')
    
#print(soup.prettify())
```

```python
### get to table and retrieve the columns
```

```python
table = soup.find('table', class_ = "wikitable sortable jquery-tablesorter")
columnList = []
for header in table.find_all('th', class_ = 'headerSort'):
    columnList.append(header.text.strip())
    
columnList
```

### get the rows and convert to panda dataframe

```python
rowmatrix = []
for row in table.find_all('tr'):
  rowdata = []
  for column in row.find_all('td'):
   # print(column.text)
    rowdata.append(column.text.strip())
  #print()
  rowmatrix.append(rowdata)
#print(rowmatrix)

# convert it to pandas datafram
df = pd.DataFrame(rowmatrix)

df.columns = columnList
df.head()
```

### clean up and perform intial data analysis

```python
dff = df.dropna()
print("df", df.shape)
print("dff", dff.shape)
dff.head(15)
```

### Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.

```python
dff = dff[dff.Borough != "Not assigned"]
dff.head(15)
```

### if a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough

```python

dff.Neighbourhood = dff.Borough.where(dff.Neighbourhood == 'Not assigned', dff.Neighbourhood)
#dff = dff[dff.Neighbourhood != "Not assigned"]
dff.head(15)
```

### visual check if the columns have any missing data

```python
print(sorted(dff.Postcode.unique()))
print(sorted(dff.Borough.unique()))
print(sorted(dff.Neighbourhood.unique()))
```

### combine neighbourhood rows with same Postcode

```python
dff = dff.groupby(['Postcode', 'Borough'])['Neighbourhood'].agg(','.join).reset_index()
dff.head(50)
```

### function that extracts the category of the venue

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

### extract venuew catgegories from json results

```python
venues = results['response']['groups'][0]['items']
#venues = results['response']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

print('{} venues were returned by Foursquare.'.format(nearby_venues.shape[0]))

nearby_venues.head()
```

### define function to get nearby venues and use it o get venues in downtown toronto borough

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)

downtownToronoto_venues = getNearbyVenues(names=dft['Neighbourhood'],
                                   latitudes=dft['Latitude'],
                                   longitudes=dft['Longitude']
                                  )
```

### Perform onehot coding on venue catgegories

```python
# one hot encoding
dft_onehot = pd.get_dummies(downtownToronoto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
dft_onehot['Neighborhood'] = downtownToronoto_venues['Neighborhood'] 

# move neighborhood column to the first column
mid = dft_onehot['Neighborhood']
dft_onehot.drop(labels=['Neighborhood'], axis=1,inplace = True)
dft_onehot.insert(0, 'Neighborhood', mid)
dft_onehot.head()
```

### group the onehot coded data

```python
dft_onehot_grouped = dft_onehot.groupby('Neighborhood').sum().reset_index()
dft_onehot_grouped.head()
```

```python
### Run kmeans cluster with 5 as cluster options
```

```python
# set number of clusters
kclusters = 5
neighborhood = dft_onehot_grouped['Neighborhood']
dft_onehot_grouped_clustering = dft_onehot_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(dft_onehot_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

### Add clutering lables to the group and also bring Neighbourhood column position to the front

```python
# add clustering labels

dft_onehot_grouped_clustering.insert(0, 'Cluster Labels', kmeans.labels_)
dft_onehot_grouped_clustering.insert(0, 'Neighbourhood', neighborhood)

dft_onehot_grouped_clustering.head()
```

### merge with the downtown tornot data frame dft

```python
dft_merged = dft.drop(labels=['Postcode', 'Borough'], axis=1)

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
dft_merged = dft_merged.join(dft_onehot_grouped_clustering.set_index('Neighbourhood'), on='Neighbourhood')

dft_merged.head() # check the last columns!
```

### visualize the clustered data

```python
import numpy as np
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(dft_merged['Latitude'], dft_merged['Longitude'], dft_merged['Neighbourhood'], 
                                  dft_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
